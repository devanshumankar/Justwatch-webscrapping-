````python
#import all necessary labraries
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
````

## **Scrapping Movies Data**

````python
def fetch_movie_urls(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
    }
    response = requests.get(url, headers=headers)
    if response.status_code != 200:
        return "Failed to retrieve the page, status code:", response.status_code
    soup = BeautifulSoup(response.text, 'html.parser')
    return soup


url = 'https://www.justwatch.com/in/movies?release_year_from=2000'
soup=fetch_movie_urls(url)
print(soup.prettify())

## Hint : Use the following code to extract the film urls
# movie_links = soup.find_all('a', href=True)
# movie_urls = [link['href'] for link in movie_links if '/movie/' in link['href']]

# url_list=[]
# for x in movie_urls:
#   url_list.append('https://www.justwatch.com'+x)
````

## **Fetching Movie URL's**

````python
# Write Your Code here
list_link=[]
for i in soup.find_all('a',attrs={'class':'title-list-grid__item--link'}):
  list_link.append("https://www.justwatch.com"+i.get('href'))
````

## **Scrapping Movie Title**

````python
pip install cloudscraper
````

````python
import cloudscraper
scraper = cloudscraper.create_scraper()
title=[]
release_year=[]
Genres=[]
IMDB_rating=[]
runtime=[]
age_rating=[]
production_country=[]
streaming_service=[]
for i in list_link:
  url=i
  response = scraper.get(url)
  soup = BeautifulSoup(response.text, 'html.parser')
  #TITLE
  try:
    for i in soup.find_all("div", attrs={"class":"title-detail-hero"}):
      title.append(i.find_all('h1',attrs={'class':"title-detail-hero__details__title"})[0].text.strip())
  except:
    title.append("NA")

  #Release year
  try:
    for i in soup.find_all("div", attrs={"class":"title-detail-hero"}):
      release_year.append(eval(i.find_all('span',attrs={'class':'release-year'})[0].text.strip()))
  except:
    release_year.append('Nill')

  #Genres
  try:
    for i in soup.find_all('div',class_='title-info title-info')[0].find_all('div',class_='detail-infos__value')[1].find_all('span'):
      Genres.append(i.text)
  except:
    Genres.append('NA')


  #imdb rating
  try:
    for i in soup.find_all('span',class_="imdb-score"):
      IMDB_rating.append(i.text.strip().split(' ')[0])
  except:
    IMDB_rating.append("Null")

  #Runtime
  try:
    for i in soup.find_all('h3',class_='detail-infos__subheading',string='Runtime')[0].find_next_sibling('div',class_='detail-infos__value'):
      runtime.append(i.text)
  except:
    runtime.append('NA')

  #age_rating
  try:
    for i in soup.find_all('h3',class_='detail-infos__subheading',string='Age rating')[0].find_next_sibling('div',class_='detail-infos__value'):
      age_rating.append(i.text)
  except:
    age_rating.append('Not listed')

 #production_country
  try:
    for i in soup.find_all('h3',class_='detail-infos__subheading',string=' Production country ')[0].find_next_sibling('div',class_='detail-infos__value'):
      production_country.append(i.text)
  except:
    production_country.append('Not listed')
````

## **Fetching Streaming Service Details**

````python
scraper = cloudscraper.create_scraper()

streaming_service = []

for url in list_link:
    response = scraper.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Streaming service
    try:
        services = [img.get('alt') for img in soup.find_all('img', class_='offer__icon')]
        streaming_service.append(services)
    except:
        streaming_service.append('NA')

print("Streaming Services:", streaming_service)
````

## **Now Creating Movies DataFrame**

````python
max_length = max(len(title), len(release_year), len(Genres), len(IMDB_rating), len(runtime), len(age_rating), len(production_country), len(streaming_service))
````

````python
def pad_list(lst, length, pad_value='Null'):
    return lst + [pad_value] * (length - len(lst))
````

````python
title = pad_list(title, max_length)
release_year = pad_list(release_year, max_length)
Genres = pad_list(Genres, max_length)
IMDB_rating = pad_list(IMDB_rating, max_length)
runtime = pad_list(runtime, max_length)
age_rating = pad_list(age_rating, max_length)
production_country = pad_list(production_country, max_length)
streaming_service = pad_list(streaming_service, max_length)
````

````python
data = {
    "Title": title,
    "Release Year": release_year,
    "Genres": Genres,
    "IMDB Rating": IMDB_rating,
    "Runtime": runtime,
    "Age Rating": age_rating,
    "Production Country": production_country,
    "Streaming Service": streaming_service
}

movie_df = pd.DataFrame(data)
````
